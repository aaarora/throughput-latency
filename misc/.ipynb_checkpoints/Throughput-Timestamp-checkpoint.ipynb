```python
import json
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matplotlob params
params = {'legend.fontsize': 'x-large',
          'figure.dpi': 100,
          'figure.figsize': (20, 5),
          'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large',
          'xtick.labelsize':'x-large',
          'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
```

```python
with open('dataPoints.json','r') as f:
    dat = json.load(f)
```

```python
len(dat)
```

```python
times = [[j['ts'] for j in i] for i in dat]
vals = [[j['val'] for j in i] for i in dat]
```

```python
def makePlot(ax,times,vals,i):
    ax.plot(np.array([datetime.datetime.fromtimestamp(j) for j in times[i]]),np.array(vals[i]) / 1073741824, label=str(i))
```

```python
fig, ax = plt.subplots()
ax.set_ylabel('Throughput (in Gbits/s)')
ax.set_xlabel('Timestamp')
ax.grid(True, which='both')
ax.set_ylim(0,100)
ax.plot(np.array([datetime.datetime.fromtimestamp(j) for j in times[0]]),np.array(vals[0]) / 1073741824, label='GridFTP',linewidth=4)
ax.plot(np.array([datetime.datetime.fromtimestamp(j) for j in times[1]]),np.array(vals[1]) / 1073741824, label='HTTPS',linewidth=4)
ax.plot(np.array([datetime.datetime.fromtimestamp(j) for j in times[2]]),np.array(vals[2]) / 1073741824, label='ethr',linewidth=4)
ax.legend()
plt.savefig('throughput.png')
```
